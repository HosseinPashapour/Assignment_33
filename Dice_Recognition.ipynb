```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

```

```python
dicecolor=cv2.imread("input/1dice.png")
dice_g=cv2.cvtColor(dicecolor,cv2.COLOR_BGR2GRAY)

dice_g[dice_g>230]=0
dice_g=cv2.blur(dice_g,(5,5))
_,dice_t=cv2.threshold(dice_g,205,255,cv2.THRESH_BINARY)

dice_c,dice_h=cv2.findContours(dice_t,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

family=dict()
for i,contour in enumerate(dice_c):
    if cv2.contourArea(contour)>120:
        if dice_h[0][i][3]==-1:
                family[i]=[]
        for j in family:
            if dice_h[0][i][3]==j:
                family[j].append(i)

for dice in family:
    if len(family[dice])>0:
        x,y,w,h=cv2.boundingRect(dice_c[dice])
        cv2.rectangle(dicecolor,[x,y],[x+w,y+h],(0,255,0),1)
        cv2.putText(dicecolor,f'{len(family[dice])}',(x+w+4,y+20),cv2.FONT_HERSHEY_COMPLEX,fontScale=0.75,color=(0,255,0),thickness=2,lineType=cv2.LINE_AA)
        for i in family[dice]:
            x,y,w,h=cv2.boundingRect(dice_c[i])
            cv2.rectangle(dicecolor,[x,y],[x+w,y+h],(0,255,0),1)

plt.imshow(cv2.cvtColor(dicecolor, cv2.COLOR_BGR2RGB))
cv2.imwrite("output/1_dice.jpg",dicecolor)
plt.axis('off')
plt.show()
```
